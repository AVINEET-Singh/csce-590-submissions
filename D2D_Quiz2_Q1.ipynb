#Data Preprocessing

##Using Sklearn(giving numbers to categories/label encoding the data)

```
#LabelBinarizer converts the string categorical variable to binary 
from sklearn.preprocessing import LabelBinarizer
lb= LabelBinarizer()
df["class"]= lb.fit_transform(df["class"])
df["own_telephone"]= lb.fit_transform(df["own_telephone"])
df["foreign_worker"]= lb.fit_transform(df["foreign_worker"])
```

```
df
```

```
#LabelEncoder converts the string categorical variable to numeric labels  
from sklearn.preprocessing import LabelEncoder
lb1= LabelEncoder()
```

```
df["checking_status"]= lb1.fit_transform(df["checking_status"])
df["credit_history"]= lb1.fit_transform(df["credit_history"])
df["purpose"]=lb1.fit_transform(df["purpose"])
df["savings_status"]= lb1.fit_transform(df["savings_status"])
df["employment"]=lb1.fit_transform(df["employment"])
df["personal_status"]= lb1.fit_transform(df["personal_status"])
df["other_parties"]= lb1.fit_transform(df["other_parties"])
df["property_magnitude"]= lb1.fit_transform(df["property_magnitude"])
df["other_payment_plans"]= lb1.fit_transform(df["other_payment_plans"])
df["housing"]= lb1.fit_transform(df["housing"])
df["job"]= lb1.fit_transform(df["job"])
```

```
df
```

##Using Pandas(one hot encoding)

```
#create columns based on the categories and remove the first column
df1 = df1.merge(pd.get_dummies(df1.checking_status, drop_first=True, prefix='checking_status'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.credit_history, drop_first=True, prefix='credit_history'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.purpose, drop_first=True, prefix='purpose'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.savings_status, drop_first=True, prefix='savings_status'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.employment, drop_first=True, prefix='employment'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.personal_status, drop_first=True, prefix='personal_status'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.other_parties, drop_first=True, prefix='other_parties'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.property_magnitude, drop_first=True, prefix='property_magnitude'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.other_payment_plans, drop_first=True, prefix='other_payment_plans'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.housing, drop_first=True, prefix='housing'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.job, drop_first=True, prefix='job'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.own_telephone, drop_first=True, prefix='own_telephone'), left_index=True, right_index=True)
df1 = df1.merge(pd.get_dummies(df1.foreign_worker, drop_first=True, prefix='foreign_worker'), left_index=True, right_index=True)
#df1 = df1.merge(pd.get_dummies(df1.class, drop_first=True, prefix='class'), left_index=True, right_index=True)
df1['class'] = df1['class'].replace(['good','bad'],['1','0'])
```

```
#deleting the parent column used for creating categorized column.
del df1["checking_status"]
del df1["credit_history"]
del df1["purpose"]
del df1["savings_status"]
del df1["employment"]
del df1["personal_status"]
del df1["other_parties"]
del df1["property_magnitude"]
del df1["other_payment_plans"]
del df1["housing"]
del df1["job"]
del df1["own_telephone"]
del df1["foreign_worker"]
```

```
#changing the order of 'class' column by copying it to a new column 
df1["class_label_Actual"]=df1["class"]
```

#Splitting Data

Splitting the dataset(created using pandas i.e., after one hot encoding) into training and test samples

```
from sklearn.model_selection import train_test_split
training_set, test_set = train_test_split(df1, test_size = 0.2, random_state = 1)
```

Classifying the predictors and target

```
X_train = training_set.iloc[:,0:48].values
Y_train = training_set.iloc[:,48].values
X_test = test_set.iloc[:,0:48].values
Y_test = test_set.iloc[:,48].values
```

#Feature Scaling

```
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import normalize 
```

```
# Scaling the data to bring all the attributes to a comparable level 
scaler = StandardScaler() 
X_scaled_train = scaler.fit_transform(X_train)
X_scaled_test = scaler.fit_transform(X_test) 
  
# Normalizing the data so that  
# the data approximately follows a Gaussian distribution 
X_normalized_train = normalize(X_scaled_train) 
X_normalized_test = normalize(X_scaled_test)
```

#Classifier 1: SVM

```
from sklearn.svm import SVC
classifier1 = SVC(kernel='rbf', random_state = 1)
classifier1.fit(X_normalized_train,Y_train)
```

```
Y_pred1 = classifier1.predict(X_normalized_test)
```

```
test_set["Predictions_with SVM"] = Y_pred1
```

```
test_set
```

```
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
print("\nAccuracy Of SVM : ", accuracy_score(Y_test,Y_pred1))

print("\nPrecision Of SVM : ", precision_score(Y_test,Y_pred1))
print("\nRecall Of SVM : ", recall_score(Y_test,Y_pred1))
print("\nF1_score Of SVM : ", f1_score(Y_test,Y_pred1))
```

```
confusion_matrix(Y_test,Y_pred1)
```

#Classifier 2: Random Forest

```
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier(max_depth=2, random_state=0)
classifier2.fit(X_normalized_train, Y_train)
```

```
Y_pred2 = classifier2.predict(X_normalized_test)
```

```
test_set["Predictions_with RF"] = Y_pred2
```

```
test_set
```

```
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
print("\nAccuracy Of RF : ", accuracy_score(Y_test,Y_pred2))

print("\nPrecision Of RF : ", precision_score(Y_test,Y_pred2))
print("\nRecall Of RF : ", recall_score(Y_test,Y_pred2))
print("\nF1_score Of RF : ", f1_score(Y_test,Y_pred2))
```

```
confusion_matrix(Y_test,Y_pred2)
```
